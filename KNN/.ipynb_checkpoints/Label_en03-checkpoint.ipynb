```python
import pandas as pd

# ----- plot tools -----
import matplotlib.pyplot as plt
import seaborn as sns

# ----- sklean -----
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import GridSearchCV, cross_val_score

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel("CarPrice.xlsx")
data.head(10)
```

```python
data.tail(10)
```

___

# Feature Engineering

## handling missing values

```python
data.dropna(inplace=True)
data.shape[0]
```

## Duplicated data

```python
data.drop_duplicates(inplace = True)
data.shape[0]
```

```python
data = data.rename(columns={"ยี่ห้อ" : "Brand", 
                            "ประเภทเชื้อเพลิง" : "Fuel_Type",
                            "ประเภทผู้ขายรถ" : "Seller_Type",
                            "ประเภทเกียร์" : "Gear_Type",
                            "กิโลเมตรที่ขับไปแล้ว" : "Mileage",
                            "ปีที่ออกขายครั้งแรก" : "First_Sell_Year",
                            "ราคาขายมือสอง" : "Secondhand_Price",
                            "จำนวนเจ้าของรถในอดีต" : "Past_Owner"
                           })
data
```

```python
data.nunique()
```

### Feature : "Brand"

```python
data.Brand.unique()
```

```python
data.Brand.str.split().str[0].unique()
```

```python
data.loc[(data["Brand"] == "Suzukii") | (data["Brand"] == "SuzukiI"), "Brand"] = "Suzuki"
```

```python
data["Model"] = data['Brand'].str.split().str[1]
data.head(5)
```

```python
data.loc[ :,'Brand'] = data['Brand'].str.split().str[0]
data.head(5)
```

```python
top_Brand = data["Brand"].str.split().str[0].value_counts().head(15)

plt.figure(figsize=(10, 6))
sns.barplot(
    x=top_Brand.values, 
    y=top_Brand.index, 
    palette="RdBu",
)
plt.title(f'Top 15 Brand')
plt.xlabel("Frequency")
plt.ylabel("Brand")
plt.tight_layout()
#
plt.savefig("Top_15_Brand.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# add new feature : Age 
data["Age"] = 2024 - data["First_Sell_Year"]

data.Past_Owner = data.Past_Owner.astype(int)
```

## Handling Outler

```python
num_cols = ["Secondhand_Price", "Mileage","Age"]
fig = plt.figure(figsize=[13,3])

for idx, col in enumerate(num_cols):  # วนลูปไปตามคอลัมน์ใน num_cols
    plt.subplot(1, 3, idx+1)  # สร้าง subplot ตามตำแหน่ง
    sns.boxplot(x=col, data=data)
    plt.title(f"Boxplot for {col}")

plt.tight_layout()
#
plt.savefig("Boxplot_feature.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
data = data[(data["Secondhand_Price"] < data["Secondhand_Price"].quantile(0.99)) &
            (data["Mileage"] < data["Mileage"].quantile(0.99))]
```

```python
data
```

___

## Endcoding

```python
le = LabelEncoder()

categorical_cols = ["Brand", "Model","Fuel_Type", "Seller_Type", "Gear_Type"]
for i in categorical_cols :
    data[i] = le.fit_transform(data[i])
data.head(5)
```

___

# Training Model

```python
# Correlation 
plt.figure(figsize=(10, 8)) 
sns.heatmap(data.corr(), annot=True, cmap="RdBu", annot_kws={"size": 10})
#plt.savefig("correlation_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Model 01

```python
feature = ["Brand", "Age", "Mileage", "Fuel_Type", "Seller_Type", "Gear_Type", "Past_Owner", "Model"]
label = "Secondhand_Price"
```

```python
X = data[feature]
y = data[label]
# ----- train test split -----
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=2)

# --- Scaler ---
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train_sub)
X_test_scaled = scaler.transform(X_test)


model_01 = KNeighborsRegressor(n_neighbors=3)
# ----- Cross-validation (k=5) -----
scores = cross_val_score(model_01, X_train_scaled, y_train_sub, cv=5, scoring="r2")
print("Cross-validation accuracy:", scores.mean())

# ----- test score -----
model_01.fit(X_train_scaled, y_train_sub)
y_pred = model_01.predict(X_test_scaled)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("R^2 : ", r2)
print("Mean Squared Error : ", mse)
print("Mean Absolute Error : ", mae)
```

# Model 02
## Use StandardScaler()

```python
feature = ["Brand", "Mileage",  "Fuel_Type", "Seller_Type", "Gear_Type", "Age"]
label = "Secondhand_Price"
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = data[feature]
y = data[label]
# ----- train test split -----
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=2)

# --- Scaler ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_sub)
X_test_scaled = scaler.transform(X_test)
model_02 = KNeighborsRegressor(n_neighbors=3)

# ----- Cross-validation (k=5) -----
scores = cross_val_score(model_02, X_train_scaled, y_train_sub, cv=5, scoring="r2")
print("Cross-validation accuracy:", scores.mean())

# ----- test score -----
model_02.fit(X_train_scaled, y_train_sub)
y_pred = model_02.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("R^2 : ", r2)
print("Mean Squared Error : ", mse)
print("Mean Absolute Error : ", mae)
```

# Model 02.1
## Use MinMaxScaler()

```python
X = data[feature]
y = data[label]
# ----- train test split -----
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=2)

# --- Scaler ---
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train_sub)
X_test_scaled = scaler.transform(X_test)
model_021 = KNeighborsRegressor(n_neighbors=3)

# ----- Cross-validation (k=5) -----
scores = cross_val_score(model_021, X_train_scaled, y_train_sub, cv=5, scoring="r2")
print("Cross-validation accuracy:", scores.mean())

# ----- test score -----
model_021.fit(X_train_scaled, y_train_sub)
y_pred = model_021.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("R^2 : ", r2)
print("Mean Squared Error : ", mse)
print("Mean Absolute Error : ", mae)
```

# Hyperparameter tuning

## tuning by Test set

```python
param_grid = {'n_neighbors': [3, 5, 7, 9, 11],
             'weights' : ['uniform', 'distance'],
             'algorithm' : ['auto','ball_tree', 'kd_tree', 'brute'],
             'metric' : ['minkowski','euclidean','manhattan']}
grid_search = GridSearchCV(KNeighborsRegressor(), param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train_sub)
print("Best n_neighbors:", grid_search.best_params_['n_neighbors'])
print("Best weights:", grid_search.best_params_['weights'])
print("Best algorithm:", grid_search.best_params_['algorithm'])
print("Best metric:", grid_search.best_params_['metric'])
```

```python
knn = KNeighborsRegressor(n_neighbors=9,weights='distance',algorithm='ball_tree',metric='manhattan')
knn.fit(X_train_scaled, y_train_sub)
y_pred = knn.predict(X_test_scaled)
test_score = r2_score(y_test, y_pred)


validation_score = knn.score(scaler.transform(X_val), y_val)
print("Validation set score:", validation_score)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print("Test set R² score:", test_score)
print("Mean Squared Error : ", mse)
print("Mean Absolute Error : ", mae)
```

```python
scores = cross_val_score(grid_search, X_train_scaled, y_train_sub, cv=5, scoring="r2")
print("Cross-validation accuracy:", scores.mean())
```
